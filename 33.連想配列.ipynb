### [概要]
連想配列の応用問題。リストを使うとかなりややこしい処理が必要だが、連想配列なら簡単に解け  
るという問題はたまに出題される。連想配列はスラスラ書けるようになろう。  

### [解説]
点の組p1(p1_x.p1_y),p2(p2_x,p2_y)を取り、それらを長方形の対角線上(左上と右下または左  
下と右上)にある点とする。図のようにp1,p2を取り、残り2点をp3,p4とすると、座標は以下のよう  
になる。  
p3=(p1_x,p2_y)  
p4=(p2_x,p1_y)  
  
p3,p4にあたる点が存在すればこの4点を頂点とする長方形が作れる。  
存在するかどうかは連想配列を使って確認する。  
p1,p2のx座標またはy座標が一致する場合は対角線にできないので、その組については考えない。  
この方法でp1,p2を全組探索するとp3,p4が存在する場合に二重に数えてしまう(左上と右下、左  
下と右上)ため、最後に答えを2で割る。  

### [実装のコツ]
<defaultdict(連想配列)>  
連想配列は別名「辞書」とも言い、「キー」と対応する「値」を登録できるデータ構造。  
pythonでは標準で搭載されており、dict()と書くことでも使える。  
しかし標準のものはデフォルトの値(初期値)が設定できず、存在しない「キー」にアクセスする可  
能性があるときの処理が面倒。  
そこで連想配列の一つ、defaultdictを使う。  
defaultdictは通常の連想配列と同様に使え、さらに初期値を設定できる。値の割当が行われて  
いない「キー」には自動的に初期値が割り振られる。  
「使い方」  
・インポート：from collections import defaultdict  
・作成(デフォルト0の場合)【O(1)】：変数名=defaultdict(int)  
・キー、値の登録【O(1)】：変数名[キー]=値  
・値の参照【O(1)】：変数名[キー]  


```python
#入力
N = int(input())
#座標格納リスト
points = []

# defaultdictを使用 初期値は0
from collections import defaultdict
p = defaultdict(int)
for i in range(N):
    x,y = map(int,input().split())
    #座標をpointsへ格納
    points.append([x,y])
    #p[(x,y)] = 1 を記録
    p[(x,y)] = 1
#答えを格納する変数
ans = 0

#p1 = 0~(N-1)
for p1 in range(N):
    #p2 = (p1+1)~(N-1)
    for p2 in range(p1+1, N):
        #p1の座標
        p1_x,p1_y = points[p1]
        #p2の座標
        p2_x,p2_y = points[p2]
        
        if p1_x == p2_x or p1_y == p2_y:
            continue
        #座標(p1_x,p2_y)と座標(p2_x,p1_y)の点が存在すれば
        if p[(p1_x,p2_y)]==1 and p[p2_x, p1_y] == 1:
            ans += 1

#二重に数えているので、2で割る
ans //= 2
print(ans)



```
