### [概要]
全探索で解けそうだが範囲を考えないといけないという問題。変数が複数ある場合はいくつかを  
固定してその取りうる範囲を考えるというのがセオリー。この考え方は他の問題でもよく使うから  
覚えておこう  
### [解説]
ABC≤Nという条件からA≤(Nの3乗根)であることがわかる。これよりAが大きければA=B=Cの場  
合でもNを超えてしまうため。  
Aが決まっている場合、Bの範囲はB≤√(N/A)となる。これよりBが大きければB=Cの場合でもNを  
超えてしまうため。  
A,Bが決定されている場合、Cの範囲はB≤C≤(N/AB)となる。よってありうるCの個数は((N/AB)の  
切り捨て-B+1))個。  
計算量はO(N^(2/3))となる。どうしてこの値が出てくるかは公式解説に記載があるが、ここまで厳  
密にわからなくてもとりあえずコードを書いて、N=10^11でコードテストをしてみればよい。  
### [実装のコツ]
<少数を避ける>  
Aの範囲はA≤(Nの三乗根)  
Bの範囲はB≤√(N/A)  
となるが右辺はどちらも少数の値となる。コンピューターは無限小数が扱えず、少数の計算を行  
うと誤差が出る可能性があるので極力避けなければならない。  
A≤(Nの三乗根)⇔A^3≤N  
B≤√(N/A)⇔A*B^2≤N  
と変換して範囲の指定を行う。 
 
<pypyで提出>  
計算量が大きく、pythonでは間に合わないので、提出の際言語に「pypy3」を選んで提出する。  
「pypy」はプログラミング言語の一つで、pythonで書いたコードを高速で実行することができる。  
(pythonは2sec以内にだいたい10^6回、pypyは2sec以内にだいたい10^8回計算が可能)  

```python
#入力を受け取る
N = int(input())
#答えのカウント
ans = 0

#初期値
A = 1
#A<=(Nの3乗根)⇔A^3<=N の間
while A**3<=N:
    #初期値
    B=A
    #B<=√(N/A)⇔A*B^2<=N の間
    while A*B**2<=N:
        #ありうるCの個数=(N/AB)の切り捨て-B+1
        ans += int(N/(A*B)) - B+1
        #次のBへ
        B += 1
    #次のAへ
    A+=1

#答えの出力
print(ans)
```
