### [概要]
問題文の指示通りに操作を行うとTLEするから工夫が必要というパターン。  
例を作る→一般化する→実装するという手順で解く。  

### [解説]
f(x)をx=1,2,3,...について計算すると以下のようになる。  
f(1)=1  
f(2)=2  
f(3)=3  
...  
f(9)=9  
f(10)=1  
f(11)=2  
f(12)=3  
f(13)=4  
...  
f(99)=90  
f(100)=1  
f(101)=2  
...  
f(999)=900  
...  
f(1)~f(N)までの和を桁を区切って計算すると以下のようになる。  
N=1~9：「1+2+...+N」  
N=10~99：「1+2+...+9」+「1+2+...+(N-10+1)」  
N=100~999：「1+2+...+9」+「1+2+...+90」+「1+2+...+(N-100+1)」  
N=1000~9999：「1+2+...+9」+「1+2+...+90」+「1+2+...+900」「1+2+...+(N-1000+1)」  
...  
丁寧に書き直すと以下のようになる  
・10^0≤N<10^1：「1~(N-10^0+1)までの和」  
・10^1≤N<10^2-1：「1~9までの和」+「1~(N-10^1+1)の和」  
・10^2≤N<10^3-1：「1~9までの和」+「1~90の和」+「1~(N-10^2+1)の和」  
・10^3≤N<10^4-1：「1~9までの和」+「1~90の和」+「1~900の和」+「1~(N-10^3+1)の和」  
上記の関係に沿って順に足し算を行っていけば答えを出せる。  
a~bまでの和は公式(B-A+1)*(A+B)/2を使って計算する。  

### [実装のコツ]
- 余りをとる  
「998244353(大きな素数)で割った余りを出力してください」というような問題はよく出題される。気  
をつけるべきなのは答えの出力直前に余りを取るのではなくて、計算の都度余りを取るというこ  
と。途中の計算で大きすぎる数の計算が発生するとTLEしてしまう恐れがある。  

```python
#余り
mod = 998244353
#入力
N = int(input())
#A~Bまでの和
#引数:A,B → 返り値:A~Bまでの和
def S(A,B):
    return (B-A+1)*(A+B)//2

#答え
ans = 0

#x=1~18
for x in range(1,19):
    if 10**x <= N:
        #「1~9*10**(x-1)までの和」
        ans += S(1,9*10**(x-1))
        #余りをとる
        ans %= mod
    #10^(x-1)<=N<10^xならば
    else:
        #「1~(N-10^(x-1)+1)までの和」
        ans += S(1, N-10**(x-1)+1)
        #余りを取る
        ans %= mod
        break

#答えを出力
print(ans)

```
