### [概要]
複雑な条件を整理するコツを身に着けよう。

条件に表記漏れがないように表形式でまとめると良い。

### [解説]
問題文の指示通りにpow(A,C),pow(B,C)を計算すると、A,B,Cが最大の10^9だった場合に桁が大きすぎてTLEする。  

pow(A,C)とpow(B,C)の大小関係は以下の4つの要素によって決まる。
- 偶数 or 奇数
- A : 0以上 (+)、0未満(-)
- B : 0以上 (+)、0未満(-)
- |A|,|B|の大小関係 : |A|<|B|,|A|=|B,|A|>|B|
全パターン考えても2**3*3=24パターンしかない。
  


### [実装のコツ]
提出ではわかりやすいようにコメントを記載し、出力をifと別の行にしている。  
しかし、自分で書くときは、このような条件分岐ではコメントなしで出力を同じ行に書いたほうが見やすい。

```
if abs(A) < abs(B):print('<')
if abs(A) == abs(B):print('=')
if abs(A) > abs(B):print('>')
```

**<絶対値>**
xの絶対値 **|x|** は

```
abs(x)
```

とかく。

また、条件がどちらでも結果が変わらないことがわかれば、省略してやってもよい。  


````python
#TLE例
#pow関数の定義
def pow(X,Y):
    return X ** Y

#A, B, Cの入力
A, B, C = map(int,input().split())

if pow(A,C) < pow(B,C):
    print('<')

elif pow(A,C) > pow(B,C):
    print('>')

else:
    print('=')

````
